```python
# –£–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ 4a: –ü–∞—Ä–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏
# TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –∫–æ–≥–¥–∞ –¥–∞—Ç–∞—Å–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω

# CODE

print("–¶–µ–ª–∏ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏—è 4a:")
print("1. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –≤—Å–µ –ø–∏—â–µ–≤—ã–µ –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ")
print("2. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ —Å–∏–ª—å–Ω—ã–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏")
print("3. –°—Ä–∞–≤–Ω–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø—Ä–æ—Ç–∏–≤ KDE –Ω–∞ –¥–∏–∞–≥–æ–Ω–∞–ª–∏")
print("4. –ù–∞–π—Ç–∏ –∫–∞–∫–∏–µ –ø–∏—Ç–∞—Ç–µ–ª—å–Ω—ã–µ –≤–µ—â–µ—Å—Ç–≤–∞ –ª—É—á—à–µ –≤—Å–µ–≥–æ —Ä–∞–∑–¥–µ–ª—è—é—Ç –æ—Ü–µ–Ω–∫–∏ Nutri-Score")

features = ['energy_100g', 'proteins_100g', 'carbohydrates_100g', 'fat_100g', 'nutriscore_grade']
df_pair = df_nutrition[features].copy()
df_pair['nutriscore_grade'] = df_pair['nutriscore_grade'].str.lower()

df_pair_sample = df_pair.sample(n=3000, random_state=42)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ pairplot
sns.set(style="whitegrid")
sns.pairplot(
    df_pair_sample,
    hue='nutriscore_grade',
    diag_kind='kde',
    palette='coolwarm',
    corner=True,
    plot_kws={'alpha': 0.5, 's': 20}
)

plt.suptitle('–ü–∞—Ä–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –ø–∏—â–µ–≤—ã—Ö –≤–µ—â–µ—Å—Ç–≤ —Å Nutri-Score', y=1.02)
plt.show()
```

```python
# –£–ø—Ä–∞–∂–Ω–µ–Ω–∏–µ 4b: –ü—Ä–æ–¥–≤–∏–Ω—É—Ç—ã–µ –º–Ω–æ–≥–æ–º–µ—Ä–Ω—ã–µ —Ç–µ—Ö–Ω–∏–∫–∏
# TODO: –†–µ–∞–ª–∏–∑—É–π—Ç–µ –∫–æ–≥–¥–∞ –¥–∞—Ç–∞—Å–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω

# –ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –ø—Ä–æ–¥–≤–∏–Ω—É—Ç—ã—Ö –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–π
# df_advanced = df_nutrition.sample(1000, random_state=42)  # –í—ã–±–æ—Ä–∫–∞ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏

# CODE
from sklearn.preprocessing import MinMaxScaler

print("–¶–µ–ª–∏ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏—è 4b:")
print("1. –ü—Ä–∏–º–µ–Ω–∏—Ç—å –ö—Ä–∏–≤—ã–µ –≠–Ω–¥—Ä—é—Å–∞ –¥–ª—è –ø–æ–∏—Å–∫–∞ –ø–∏—â–µ–≤—ã—Ö –ø–∞—Ç—Ç–µ—Ä–Ω–æ–≤")
print("2. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ü–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –ø—Ä–æ—Ñ–∏–ª–µ–π")
print("3. –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—é RadViz –ø–æ—Ö–æ–∂–∏—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤")
print("4. –°—Ä–∞–≤–Ω–∏—Ç—å –≤—ã–≤–æ–¥—ã –æ—Ç –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥–≤–∏–Ω—É—Ç–æ–≥–æ –º–µ—Ç–æ–¥–∞")


# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
cols = ['energy_100g', 'proteins_100g', 'carbohydrates_100g', 'fat_100g', 'sugars_100g', 'sodium_100g', 'nutriscore_grade']
df_advanced = df_nutrition[cols].dropna().copy()
df_advanced = df_advanced.sample(n=1000, random_state=42)
df_advanced['nutriscore_grade'] = df_advanced['nutriscore_grade'].str.lower()

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è
scaler = MinMaxScaler()
df_advanced_scaled = df_advanced.copy()
df_advanced_scaled[df_advanced.columns[:-1]] = scaler.fit_transform(df_advanced[df_advanced.columns[:-1]])

# –ö—Ä–∏–≤—ã–µ –≠–Ω–¥—Ä—é—Å–∞
plt.figure(figsize=(10, 6))
andrews_curves(df_advanced_scaled, 'nutriscore_grade', colormap='cool')
plt.title('–ö—Ä–∏–≤—ã–µ –≠–Ω–¥—Ä—é—Å–∞: –ü–∏—â–µ–≤—ã–µ –ø—Ä–æ—Ñ–∏–ª–∏ –ø–æ Nutri-Score')
plt.grid(True)
plt.show()

# –ü–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã
plt.figure(figsize=(12, 6))
parallel_coordinates(df_advanced_scaled, 'nutriscore_grade', colormap='Set1', alpha=0.4)
plt.title('–ü–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã: –ü–∏—â–µ–≤—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã –ø–æ Nutri-Score')
plt.grid(True)
plt.show()

# RadViz
plt.figure(figsize=(8, 8))
radviz(df_advanced_scaled, 'nutriscore_grade', colormap='tab10')
plt.title('RadViz: –ö–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –ø–∏—â–µ–≤—ã—Ö –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–æ Nutri-Score')
plt.grid(True)
plt.show()
```

```python
print("–í—ã–∑–æ–≤ 4")
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)

# –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ df_clean –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω –∏ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω—É–∂–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã

app.layout = html.Div([
    html.H1('–ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –ø–∏—Ç–∞–Ω–∏—è'),

    dcc.Dropdown(
        id='category-dropdown',
        options=[{'label': str(c), 'value': str(c)} for c in sorted(df_clean['main_category'].dropna().astype(str).unique())],
        placeholder="–í—ã–±–µ—Ä–∏—Ç–µ –∫–∞—Ç–µ–≥–æ—Ä–∏—é",
        multi=True
    ),

    dcc.Dropdown(
        id='nutriscore-dropdown',
        options=[{'label': str(c).upper(), 'value': str(c)} for c in sorted(df_clean['nutriscore_grade'].dropna().astype(str).unique())],
        placeholder="–í—ã–±–µ—Ä–∏—Ç–µ Nutri-Score",
        multi=True
    ),

    dcc.Graph(id='scatter-plot')
])


@app.callback(
    Output('scatter-plot', 'figure'),
    Input('category-dropdown', 'value'),
    Input('nutriscore-dropdown', 'value')
)
def update_graph(selected_categories, selected_nutriscore):
    dff = df_clean.copy()
    if selected_categories:
        dff = dff[dff['main_category'].isin(selected_categories)]
    if selected_nutriscore:
        dff = dff[dff['nutriscore_grade'].isin(selected_nutriscore)]
    
    fig = px.scatter(dff, x='energy_100g', y='proteins_100g',
                     color='nutriscore_grade',
                     hover_data=['product_name', 'main_category'],
                     labels={'energy_100g': '–≠–Ω–µ—Ä–≥–∏—è (–∫–∫–∞–ª/100–≥)', 'proteins_100g': '–ë–µ–ª–∫–∏ (–≥/100–≥)', 'nutriscore_grade': 'Nutri-Score'},
                     title="–≠–Ω–µ—Ä–≥–∏—è vs –ë–µ–ª–∫–∏ –≤ –ø—Ä–æ–¥—É–∫—Ç–∞—Ö")
    return fig

if __name__ == '__main__':
    app.run(debug=True)
```
